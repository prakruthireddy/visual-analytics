```
import pandas as pd
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.palettes import Spectral3
from bokeh.colors import groups

output_file('trend.html')
# For Google Colab Notebook
output_notebook()

# Reading data from the .csv file
df = pd.read_csv('lava.csv')
df['Time'] = pd.to_datetime(df['Time'], format='%m/%d/%Y')
group = df.groupby(pd.Grouper(key='Time', freq='M'))['Mind', 'Shrek', 'Fidget', 'SaltBae'].sum()
group = group/1
source = ColumnDataSource(group)

# Plotting the graph
p = figure(x_axis_type="datetime", plot_width=1000, plot_height=500)
p.line(x='Time', y='Shrek', line_width=2, source=source, color='mediumblue', muted_color='aliceblue', legend='Shrek Meme')
p.line(x='Time', y='Fidget', line_width=2,color='darkolivegreen', muted_color='honeydew', source=source, legend='Fidget Spinner')
p.line(x='Time', y='SaltBae', line_width=2, source=source, color='hotpink', muted_color='mistyrose', legend='Salt Bae')
p.line(x='Time', y='Mind', line_width=2, source=source, color='orangered', muted_color='cornsilk', legend='Change My Mind')

# Giving the graph a title and changing its size
p.title.text = 'Search Interest Over Time (2016 - 2018) for Various Memes'
p.title.text_font_size = '24.5pt'

# Labelling and Modifying the axis labels' size
p.xaxis.axis_label = 'Time frame'
p.xaxis.axis_label_text_font_size = "18pt"
p.yaxis.axis_label = 'Trend'
p.yaxis.axis_label_text_font_size = "18pt"

# Modifying the legend's appearance
p.legend.location = "top_left"
p.legend.click_policy="mute"
p.legend.label_text_font_size = '12pt'

show(p)

# NOTE
# I have made the legend interactive!
# Clicking or tapping on the legend entries will hide or mute the corresponding glyph in a plot.
```
